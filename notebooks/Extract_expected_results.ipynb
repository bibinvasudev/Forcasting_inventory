```python
import glob
import os
from shutil import copyfile
```

```python
from pathlib import Path
path = Path("/Users/ow/Desktop")
output_folder = "expected_results"
date = "202003"
```

```python
os.chdir(path)
os.mkdir(output_folder)
model_paths = glob.glob(f"./backward-run/08 Predictions/*")
dirs = [sorted(glob.glob(f"{model_path}/{date}/*"))[-1] for model_path in model_paths]
accounts = [mp.replace("./backward-run/08 Predictions/Forecast ", "") for mp in model_paths]
```

```python
for dir, acc in zip(dirs, accounts):
    copyfile(f"{dir}/result_data.csv", f"{output_folder}/WA_{acc}_result_data.csv")
```
